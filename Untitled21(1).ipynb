```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/sample_data/heart.csv")

print("---FIRST FIVE ROWS OF DATA---")
print(df.head())
print("\n---EMPTY VALUES PER COLUMN---")
print(df.isnull().sum())
df.drop_duplicates(inplace=True)
df.columns = df.columns.str.strip().str.lower().str.replace('','_')
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
print("\n--- DATA CLEANING COMPLETED! ---")
print("cleaned Columns:", df.columns.tolist())
df.to_csv('cleaned_dataset.csv', index=False)
print("\nSaved clean data as 'cleaned_dataset.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('cleaned_dataset.csv')
sns.set_theme(style="darkgrid")
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Heart Disease Dataset Analysis Dashboard', fontsize=22, fontweight='bold', color='#2c3e50')
sns.histplot(data=data, x=data.columns[0], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Age Distribution of Patients', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Age')
sns.scatterplot(data=data, x=data.columns[4], y=data.columns[7], hue=data.columns[-1], palette='Set2', ax=axes[0, 1])
axes[0, 1].set_title('Cholesterol vs. Max Heart Rate Achieved', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Cholesterol')
axes[0, 1].set_ylabel('Max Heart Rate (thalach)')
sns.countplot(data=data, x=data.columns[-1], ax=axes[1, 0], palette='pastel')
axes[1, 0].set_title('Count of Patients (0 = No Disease, 1 = Disease)', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Target Status')
axes[1, 0].set_ylabel('Number of Patients')
sns.barplot(data=data, x=data.columns[1], y=data.columns[4], ax=axes[1, 1], palette='muted', errorbar=None)
axes[1, 1].set_title('Average Cholesterol Levels by Sex', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Sex (0 = Female, 1 = Male)')
axes[1, 1].set_ylabel('Avg Cholesterol')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('dashboard_output.png', dpi=300)
plt.show()
```
